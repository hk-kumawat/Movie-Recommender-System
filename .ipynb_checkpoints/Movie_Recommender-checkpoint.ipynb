# Movie Recommendation System

This notebook builds an advanced movie recommendation system using the MovieLens 1M dataset, integrating collaborative filtering and content-based filtering with genre-based adjustments. Users can enter a movie title (case-insensitive) and the desired number of recommendations. Recommendations are tailored by genres and calculated with a weighted similarity approach.

## Step 1: Importing Libraries
We start by importing essential libraries for data processing, web requests, and cosine similarity calculations. Additionally, libraries like `CountVectorizer` help in genre-based filtering.


```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
import requests
import zipfile
from io import BytesIO
```

## Step 2: Downloading and Extracting the Dataset
We download the MovieLens 1M dataset from the web, which contains information on movies, ratings, and genres. The downloaded zip file is then extracted for easy access to its contents.


```python
# Download and unzip MovieLens 1M dataset
url = "http://files.grouplens.org/datasets/movielens/ml-1m.zip"
response = requests.get(url)
with zipfile.ZipFile(BytesIO(response.content)) as z:
    z.extractall("/content/Dataset")
```

## Step 3: Loading Data Files
After extracting, we load the movie and ratings data into Pandas dataframes. We set custom column names for better readability.


```python
# Load the data files
movies = pd.read_csv('/content/Dataset/ml-1m/movies.dat', sep='::', names=['movie_id', 'title', 'genres'], engine='python', encoding='latin-1')
ratings = pd.read_csv('/content/Dataset/ml-1m/ratings.dat', sep='::', names=['user_id', 'movie_id', 'rating', 'timestamp'], engine='python', encoding='latin-1')
```

```python
# Plot the distribution of genres in the MovieLens dataset.
genre_counts = movies['genres'].str.get_dummies(sep='|').sum().sort_values(ascending=False)
genre_counts.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Genre Distribution in MovieLens Dataset')
plt.xlabel('Genres')
plt.ylabel('Number of Movies')
plt.show()
```

## Step 4: Merging Data
We merge the movies and ratings datasets to create a unified dataframe, making it easier to compute user-based and genre-based similarities.


```python
# Merge movies and ratings
movie_ratings = pd.merge(ratings, movies, on='movie_id')
```

## Step 5: Creating User-Item and Genre Matrices
To build a recommendation model, we generate two matrices:
1. **User-Item Matrix**: For collaborative filtering, where rows represent users, columns represent movies, and values are user ratings.
2. **Genre-Based Matrix**: Using a count vectorizer to create a matrix where movies are represented by their genres, allowing for genre-based filtering.


```python
# User-item matrix for collaborative filtering
user_movie_matrix = movie_ratings.pivot_table(index='user_id', columns='title', values='rating').fillna(0)

# Genre-based matrix
count_vectorizer = CountVectorizer(tokenizer=lambda x: x.split('|'))
genre_matrix = count_vectorizer.fit_transform(movies['genres'])
genre_sim_matrix = cosine_similarity(genre_matrix)
genre_sim_df = pd.DataFrame(genre_sim_matrix, index=movies['title'], columns=movies['title'])
```

## Step 7: Defining Helper Functions
These helper functions enable:
1. **Partial Title Matching**: Finds movies based on partial titles.
2. **Genre Filtering**: Matches genres for more personalized recommendations.


```python
# Helper function for case-insensitive, partial matching of movie titles
def find_closest_title(input_title):
    lower_titles = movies['title'].str.lower()
    matches = movies[lower_titles.str.contains(input_title.lower(), na=False)]
    if not matches.empty:
        return matches.iloc[0]['title']  # Return the closest match found
    else:
        return None


# Helper function to filter movies by genre
def get_genres(movie_title):
    genres = movies[movies['title'] == movie_title]['genres']
    if not genres.empty:
        return genres.values[0].split('|')
    return []
```

## Step 8: Recommendation Function
This function provides recommendations by finding similar movies based on collaborative and genre-based filtering, then filtering for similar genres.


```python
# Update recommendation function to filter by genres
def recommend_movies(movie_title, num_recommendations=5):
    # Find the closest matching title
    exact_title = find_closest_title(movie_title)
    if not exact_title:
        return f"Movie '{movie_title}' not found in dataset. Please try another title."

    # Get genres of the input movie
    input_genres = get_genres(exact_title)

    # Enforce recommendation limit
    num_recommendations = min(num_recommendations, 10)

    # Fetch similarity scores
    try:
        similar_movies = combined_sim_df[exact_title].sort_values(ascending=False)
    except KeyError:
        return f"Could not find similar movies for '{exact_title}'."

    # Filter recommended movies by matching genres
    genre_filtered = similar_movies.index[similar_movies.index.isin(movies[movies['genres'].apply(lambda g: any(genre in g for genre in input_genres))]['title'])]
    recommended_movies = similar_movies.loc[genre_filtered].drop(exact_title).head(num_recommendations)

    # Format output with similarity scores
    recommendations = pd.DataFrame({
        'Recommended Movie': recommended_movies.index,
        'Similarity Score': recommended_movies.values
    }).reset_index(drop=True)

    return recommendations
```

## Step 9: Testing the Recommendation System
Input a movie title and the desired number of recommendations to see the results. The model will suggest movies similar in both genre and collaborative filtering.


```python
movie_title = input("Enter a movie title: ")
num_recommendations = int(input("Enter the number of recommendations (max 10): "))

recommendations = recommend_movies(movie_title, num_recommendations)
print(f"\nRecommendations for '{movie_title}':\n")
print(recommendations)

```

```python
# Show similarity scores for the top recommended movies for the input movie.

plt.figure(figsize=(10, 6))
plt.barh(recommendations['Recommended Movie'], recommendations['Similarity Score'], color='salmon')
plt.xlabel('Similarity Score')
plt.ylabel('Recommended Movies')
plt.title(f'Similarity Scores of Top Recommended Movies for "{movie_title}"')
plt.gca().invert_yaxis()
plt.show()
```
